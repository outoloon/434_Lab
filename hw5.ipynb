```python
fig, ax = plt.subplots(1, 1)
#ax.hist(scipy.stats.poisson.rvs(X, size=100000),50, density=True)
mean, var, skew, kurt = poisson.stats(100, moments='mvsk')
#plt.yscale('log')
plt.tick_params(labelsize = 10)
plt.title('Simulated Background + injected signal', fontsize='15')
plt.xlabel('Signal Strength', fontsize='15')
plt.ylabel('# of events', fontsize='15')
b = stats.norm.rvs(0,3,size = 1000000)
d = np.linspace(0,20,1000000)
ax.hist(d,186,alpha = 0.6, label = 'Background')
ax.hist(b,186,alpha = 0.6, label = 'Injected Signal')
ax.hist(b+d,186,alpha = 0.6, label = 'Observed Data')
plt.legend()
plt.show()

data = b + d

signaledges = np.linspace(0,20,40)
dataedges = np.linspace(-7,27,68)

Psd, temp, temp2= np.histogram2d(data,d, bins=[dataedges,signaledges],density=True)

datacenters = (dataedges[:-1] + dataedges[1:]) / 2
signalcenters = (signaledges[:-1] + signaledges[1:]) / 2
plt.pcolormesh(datacenters,signalcenters,Psd.T)
plt.ylabel('True signal, $P(s|d)$', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
plt.show()
data1 = b + 15

plt.ylabel('Probability', fontsize = 24)
plt.xlabel('Observed data, $P(d|s)$', fontsize = 24)
plt.hist(data1, 100,density=True)
plt.show()
data2 = 15-b

plt.ylabel('probability', fontsize = 24)
plt.xlabel('True signal $P(s|d)$', fontsize = 24)
plt.hist(data2, 100,density=True)
plt.show()
```

1b)  After slecting a signal of 15, we can look at the slice of data (the 3rd graph above  -  probability vs p(d|s) ). We are essentialy taking a horizontal line of data from the 2d histogram at the y value of 15. comparing the resulting distrabution to the heat map of the 2d distrabution i am not suprised it looks like  gaussian that has been shifted up from zero by 15. Since our observed data is just a background gaussian plus a certain integer. 
This graph tells us the probability distrabution of what we will observe given a particular real signal. In this case it is not suprising that 15 (the real signal) is the most likely result since the background is a gaussian centered at zero, making the most likely observed data point 15 + 0 = 15.

1c) Going in the opposite direction, taking a vertical slice of our 2d histgram, we get an identical guassian showing us the probability distrabution of true signals that coul have given us an observed signal of 15. It is not supprising that these are identical guassians since the background is centered at zero, and we picked an identical point ofr our obsevred data and our real signal.

Changing nothing but the mean value of the background distrabution, we see that our histograms look very similiar but are all shifted over, some shifted a different amount than the others. In particular, our observed data probability distrabution is shifted up the same amount as the background. This makes sense, our observed data is just background+signal. The true signal probability distrabution on the other hand is actually shifted down. This makes total sense algebraicly. The second histogram is showing us what the real data could be if the observed data was equal to 15 or   { observed = 15 = signal + background } which rearranges to be {signal = 15 - background} which would give us a background guassian shifted down from 15. 

notably, these two guassians meet in the middle at 15, the point we were using to generate them.
