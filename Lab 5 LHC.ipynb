1.Do all features provide discrimination power between signal and background?
    
Some parameters discriminate between signal and background very dramaticly others are almost indistinguishable and therefor not very usful to us. 

Some notable parameters include: 

Mass and pt, both of which have a very distinct spike in the signal data that looks nothing like the background. 

ee2,t2,t3,t21,ktdeltaR all have fairly differentiable signals but not to the degree that mass and pt are differentiable. These ar characterized more by a shift in the weight of the distrabution more than any change in amplitude. 

eta,phi,d2,ee3,angularity,t1, and t32 all have signals that look a bit too much like the background distrabution to be very useful, some signals (like ee3) dont even make it out of the background distrabutions amplitude or range.

2.Are there correlations among these features?

3. below are computations for the discovery senstivity for a mass and pt measurement.

To begin, on the left there is a plot of both signal and background. I used this to pick a small x-range in which the background and signal vary significantly different. This can be seen in the second (middle) plot. I limited the histogram x axis to 483-486, and set the bins to 1, which gave me the total number of events in that region. Since this is simmulated data, I assumed it was a resonable average event number. I used the total background events in the range 483-486 as my average when caluculting and plotting my poisson. Using this poisson I found the 5 sigma limit. Which in the case of pt (in the range 483-486) was 2544 events. 

```python
plt.subplot(2, 3, 1)
plt.hist(new_dict['pt'],500,histtype = 'step')
plt.hist(higg['pt'],500,histtype = 'step')
plt.title('Pt')
plt.subplot(2, 3, 2)
plt.hist(new_dict['pt'],50,alpha=0.8,range = (483,486))
plt.hist(higg['pt'],50,alpha=0.5,range = (483,486))
plt.title('Pt - sliced range')
plt.xlim(483,486)
plt.subplot(2, 3, 3)
plt.hist(new_dict['pt'],1,histtype = 'step',range = (483,486))
plt.hist(higg['pt'],1,histtype = 'step',range = (483,486))
plt.title('Pt - total events within range')
plt.xlim(483,486)
plt.show()
fig, ax = plt.subplots(1,1)
mu = 2300
x = np.arange(poisson.ppf(0.000001, mu),poisson.ppf(0.999999, mu))
ax.plot(x, poisson.pmf(x, mu))
plt.show()
sig = poisson.ppf((norm.cdf(5)),mu)
print("5 sigma detection boundry =",sig)
```

Below I did the same thing as above, but with mass. Finding a 5 sigma boundry of about total 6900 events within the range 120-132.

```python
plt.subplot(2, 3, 1)
plt.hist(new_dict['mass'],500,histtype = 'step')
plt.hist(higg['mass'],500,histtype = 'step')
plt.title('mass')
plt.subplot(2, 3, 2)
plt.hist(new_dict['mass'],50,alpha=0.8,range = (120,132))
plt.hist(higg['mass'],50,alpha=0.5,range = (120,132))
plt.title('mass - sliced range')
plt.xlim(120,132)
plt.subplot(2, 3, 3)
plt.hist(new_dict['mass'],1,histtype = 'step',range = (120,132))
plt.hist(higg['mass'],1,histtype = 'step',range = (120,132))
plt.title('mass - total events within range')
plt.xlim(120,132)
plt.show()
fig, ax = plt.subplots(1,1)
mu = 6500
x = np.arange(poisson.ppf(0.000001, mu),poisson.ppf(0.999999, mu))
ax.plot(x, poisson.pmf(x, mu))
plt.show()
sig = poisson.ppf((norm.cdf(5)),mu)
print("5 sigma detection boundry =",sig)
```
